```python
from sklearn import linear_model
import pandas as pd
import numpy as np

house_price = linear_model.LinearRegression()

n,m = [int(var) for var in input().split()]
train = pd.DataFrame(np.zeros((m, n+1)))
for i in range(m):
    train.iloc[i]=[float(var) for var in input().split()]
train_x = train.iloc[:,:n]
train_y = train[n]
l = int(input())
test = pd.DataFrame(np.zeros((l, n)))
for j in range(l):
    test.iloc[j]=[float(var) for var in input().split()]

house_price.fit(train_x, train_y)
res = house_price.predict(test)

print(*res, sep='\n')
```
